## Question 1 - Fun with Fixed Point iteration

Let's say that we wanted to calculate $\sqrt{M}$ given that $M \in \mathbb{R}$ and $M > 0$ and that we did not want to use the function `sqrt` directly.  Consider the function 

$$
    f(x) = x^2 - M
$$

which has a positive root at at $x=\sqrt{M}$.   Here we will try to solve this using  fixed point iteration

$$
    x_{k+1} = g(x_k)
$$
for several different iteration functions $g(x)$ (see below)

For each function $g$:

- Confirm analytically that $x=\sqrt{M}$ is a fixed point of $g(x)$
- Analyze the convergence behavior close to the fixed-point and determine the stability of the iterative scheme
- Now set $M=3$  and $x_0 = 2$.  Make a nicely labeled convergence plot showing the log of the relative error as a function of iteration. How many iterations are required to reduce the relative error to machine precision (if it converges)?

**Remember**: if you want to write $g^\prime$, remember to use `g^\prime` rather than `g'` in $\LaTeX$

**(a)** [4] To help with the last part,  write a short python function to return N iterations of a fixed point iteration.

```python
def fixed_point(g, x0, N=5):
    """conduct N iterations of the fixed point iteration x_{k+1} = g(x_k)
    
    :Input:
        - *g* (function) - The iteration function
        - *x_0* (float) - Initial iterate
        - *N* (int) - number of terms in iteration
     
    :Output:
        x: numpy array (float)
            array containing the iterations [ x0, x1, ... x_N ]
    """
    # YOUR CODE HERE
    x=[]
    x_n = x0
    x.append(x0)
    for n in range(N):
        x_n = g(x_n)
        x.append(x_n)
    x = numpy.array(x)
    return x
#raise NotImplementedError()
```

**(b)** [6] analyze the problem with 
$$ g(x) = 1 + x - \frac{x^2}{M}$$

YOUR ANSWER HERE

- **Confirm analytically that $x=\sqrt{M}$ is a fixed point of $g(x)$**

If $x=\sqrt{M}$ is a fixed point of $g(x)$,

then $g(x) - x = 0$ when $x=\sqrt{M}$

$g(\sqrt{M}) - \sqrt{M} = 0$

by substitution,

$[1 + \sqrt{M} - \frac{\sqrt{M}^2}{M}] - \sqrt{M} = 0$

$[1 + \sqrt{M} - \frac{|M|}{M}] - \sqrt{M} = 0$
 
Since we know that M > 0, $-\frac{|M|}{M} = -1$

$[1 + \sqrt{M} - 1] - \sqrt{M} = 0$

$[\sqrt{M}] - \sqrt{M} = 0$

Which evaluates to $0 = 0$

This holds. Therefore, $x=\sqrt{M}$ is shown analytically to be a fixed point of $g(x)$

- **Analyze the convergence behavior close to the fixed-point and determine the stability of the iterative scheme**

To analyze the convergence behavior close to the fixed point, we take the derivative of g(x) at the fixed point, and evaluate if it is < 1 (convergent) or > 1 (divergent).

We have $g(x) = 1 + x - \frac{x^2}{M}$ with $x^* \approx \sqrt{M}$

$|g^\prime (x^*)| = |1-\frac{2x}{M}|$

$|g^\prime (\sqrt{M})| \approx |1-\frac{2\sqrt{M}}{M}|$

$\approx |1-\frac{2}{\sqrt{M}}|$

If $|1-\frac{2}{\sqrt{M}}| < 1$ then the fixed point is stable (convergence).

So we can evaluate $|1-\frac{2}{\sqrt{M}}| < 1$

$-1 < 1-\frac{2}{\sqrt{M}} < 1$

Left inequality: $-1 < 1-\frac{2}{\sqrt{M}}$

$-2 < -\frac{2}{\sqrt{M}}$

$1 > \frac{1}{\sqrt{M}}$

$\sqrt{M} > 1$

$M > 1$

Right inequality: $1-\frac{2}{\sqrt{M}} < 1$

$-\frac{2}{\sqrt{M}} < 0$

$\frac{2}{\sqrt{M}} > 0$


Since M > 0, $\frac{2}{\sqrt{M}} > 0$ is true for all M.

So $|1-\frac{2}{\sqrt{M}}| < 1$ always holds true when M > 1, to fulfill the left inequality. Therefore the iterative scheme is stable (converges) when M > 1, otherwise it diverges and exhibits instability when M < 1.

- **Now set $M=3$  and $x_0 = 2$.  Make a nicely labeled convergence plot showing the log of the relative error as a function of iteration. How many iterations are required to reduce the relative error to machine precision (if it converges)?**

20 iterations are required to reduce the relative error to machine precision.

```python
# Place plotting code here if needed

# YOUR CODE HERE
M = 3
x0 = 2
g = 1 + sympy.symbols('x') - (sympy.symbols('x')**2) / M
g = sympy.lambdify(sympy.symbols('x'), g, modules=['numpy'])
x = fixed_point(g, x0, 1000)
true = numpy.sqrt(M)

print("True value to approximate:", true)
eps = numpy.finfo(float).eps
print("Machine precision:", eps, "\n")

num_steps = 0
log_relative_error_list = []
print('\tIteration\t      x\t      relative error\t      numpy.log(relative error)')
for steps in range(len(x)):
    relative_error = numpy.abs(true-x[steps])/numpy.abs(true)
    print("{:12}\t{:12.16f}\t{:12.16f}\t{:12.16f}".format(num_steps+1, x[steps], relative_error, numpy.log(relative_error)))
    num_steps += 1
    log_relative_error_list.append(numpy.log(relative_error))
    if relative_error <= eps:
        print("\nDONE:", num_steps, "iterations to reduce relative error to machine precision.")
        break

num_steps_list = numpy.arange(num_steps)
print('\nLog of relative errors:', log_relative_error_list, '\n')
fig = plt.figure(figsize=(6,6))
plt.plot(num_steps_list, log_relative_error_list, 'r',label='Log of relative error')
#plt.plot(num_steps_list, num_steps_list, 'b',label='$Iterations$')
plt.xlabel("Number of Iterations",fontsize=16)
plt.ylabel("Log of Relative Error: numpy.log()",fontsize=16)
plt.legend(fontsize=14)
plt.grid()
plt.title('Convergence plot: Iterations v.s. Log of Relative Error')
plt.show()    

#raise NotImplementedError()
```

**(d)** [6] Newton's method
$$ g(x) = x - \frac{f}{f^\prime}$$

with $$ f = x^2 - M$$

YOUR ANSWER HERE

- **Confirm analytically that $x=\sqrt{M}$ is a fixed point of $g(x)$**

If $x=\sqrt{M}$ is a fixed point of $g(x)$, then $g(x) - x = 0$ when $x=\sqrt{M}$

$g(\sqrt{M}) - \sqrt{M} = 0$

by substitution,

$[\sqrt{M} - \frac{f(\sqrt{M})}{f^\prime(\sqrt{M})}] - \sqrt{M} = 0$

$[\sqrt{M} - \frac{\sqrt{M}^2-M}{2\sqrt{M}}] - \sqrt{M} = 0$

$[\sqrt{M} - \frac{M-M}{2\sqrt{M}}] - \sqrt{M} = 0$

$[\sqrt{M} - 0] - \sqrt{M} = 0$

This holds. Therefore, $x=\sqrt{M}$ is shown analytically to be a fixed point of $g(x)$

- **Analyze the convergence behavior close to the fixed-point and determine the stability of the iterative scheme**

To analyze the convergence behavior close to the fixed point, we take the derivative of g(x) at the fixed point, and evaluate if it is < 1 (convergent) or > 1 (divergent).

We have $g(x) = x - \frac{f}{f^\prime}$ with $x^* \approx \sqrt{M}$

$g(x) = x - \frac{x^2-M}{2x} = x - \frac{1}{2}x + \frac{M}{2}x^{-1}$

$|g^\prime (x^*)| = |1-\frac{1}{2}-\frac{M}{2}x^{-2}| = |\frac{1-M}{2x^2}|$

$|g^\prime (\sqrt{M})| \approx |\frac{1-M}{2M}|$

We can evaluate for when $|\frac{1-M}{2M}| < 1$. When this holds true, the iterative scheme is stable.

$-1 < \frac{1-M}{2M} < 1$

Left inequality:

$-1 < \frac{1-M}{2M}$

$-2M < 1-M$

$-M < 1$

$M > -1$

Since we are given that M > 0, this always holds true.

Right inequality:

$\frac{1-M}{2M} < 1$

$1-M < 2M$

$1 < 3M$

$\frac{1}{3} < M$ or $M > \frac{1}{3}$

Therefore the iterative scheme is stable when M is greater than $\frac{1}{3}$ (converges) and unstable otherwise (diverges).

For each routine,  report

* the value of $f$ at the root $f(x^*$)

```python
# YOUR CODE HERE
from  rootfinding_optimization_schemes import *
from scipy.optimize import root_scalar
params = [ (0., 0.), (4., 5.),  (27, 50.) ]
A = 0.05
x0, x1 = 0.2, 2.

counter = 1

def f_brentq(x, A, a, B):
    return numpy.cos(x) + A*(numpy.sin(a*x) + numpy.sin(B*x))

for param in params:
    print(counter, ': Given (alpha, Beta) =', param, ':')
    f = lambda x: numpy.cos(x) + A*(numpy.sin(param[0]*x) + numpy.sin(param[1]*x))
    f_prime = lambda x: -numpy.sin(x) + A * (param[0] * numpy.cos(param[0]*x) + param[1] * numpy.cos(param[1]*x))

    x_star, x_array = bisection(f, x0, x1)
    print('Bisection: Root at x = {}, f(x*) = {}, {} iterations'.format(x_star,f(x_star),len(x_array)))
    
    x_star, x_array = newton(f, f_prime, x0, tol=1.e-6)
    print('Newton: Root at x = {}, f(x*) = {}, {} iterations'.format(x_star,f(x_star),len(x_array)))
    
    x_star, x_array = secant(f, x0, x1)
    print('Secant: Root at x = {}, f(x*) = {}, {} iterations'.format(x_star,f(x_star),len(x_array)))
    
    try:
        sol = root_scalar(f_brentq, args=(A, param[0], param[1]), bracket=(x0, x1), method='brentq')
        print('brentq: Root at x = {}, f(x*) = {}, {} iterations'.format(sol.root, f(sol.root), sol.iterations))
        
        sol = root_scalar(f_brentq, args=(A, param[0], param[1]), bracket=(x0, x1), method='ridder')
        print('ridder: Root at x = {}, f(x*) = {}, {} iterations'.format(sol.root, f(sol.root), sol.iterations))
        
    except Exception as e:
        print('Root scaler failed due to error:')
        print(e)

    counter+=1
#raise NotImplementedError()
```

```python
def newtsafe(f,f_prime,a,b,tol = 1.e-6, max_steps=1000):
    """ This function implements a hybrid root-finding function that uses Newton's method and bisection.
        It tries to find a root of a given function f within an initial bracket [a, b]
        by alternating between Newton's method and bisection method to make sure the root remains
        within the bracket.
        
    Parameters:
    -----------
    
    f: function (float)
        function returning f(x)
    f_prime:  function (float)
        function returning derivative of f with respect to x
    a: float
        left bracket a < x
    b: float
        right bracket x < b
        
        Note:  the signs of f(a) and f(b) must be different to insure a bracket
    tol: float
        tolerance.  Returns when |f(x)| < tol
    max_steps: int
        maximum number of iteration steps
        
    Returns:
    --------
    x: float
        midpoint of final bracket
    n_newt:  number of newton steps for successful newton iteration
    n_bisection: number of bisection steps
        
    Raises:
    -------
    ValueError:  
        if initial bracket is invalid 
    Warning: 
        if number of iterations exceed MAX_STEPS
    Prints warning if Newton exceeds the bracket
    
    """
# YOUR CODE HERE
    MAX_STEPS = max_steps
    
    # Check bracket
    if numpy.sign(f(a)) == numpy.sign(f(b)):
        raise ValueError("Invalid initial bracket: f(a) and f(b) must have different signs")
    
    # Initialize variables
    n_newt = 0
    n_bisection = 0
    x_mid = (a + b) / 2
    
    # Newton's method from the midpoint
    x_newt = x_mid
    for n_newt in range(1, MAX_STEPS + 1):
        x_newt = x_newt - f(x_newt) / f_prime(x_newt)
        
        # Check if Newton iteration goes out of bracket
        if x_newt < a or x_newt > b:
            warnings.warn("Warning: Newton's method exceeded the bracket, switching to bisection")
            break
        
        # Check tolerance
        if numpy.abs(f(x_newt)) < tol:
            return x_newt, n_newt, n_bisection
        
    # When Newton's method fails or exceeds bracket, use bisection
    while n_bisection < MAX_STEPS:
        x_mid = (a + b) / 2
        if numpy.abs(f(x_mid)) < tol:
            return x_mid, n_newt, n_bisection
        
        # Update bracket based on sign change
        if numpy.sign(f(a)) != numpy.sign(f(x_mid)):
            b = x_mid
        else:
            a = x_mid
        
        n_bisection += 1
    
    warnings.warn('Maximum number of steps exceeded')
    return x_mid, n_newt, n_bisection
#raise NotImplementedError()
```
